```python
import json
import matplotlib.pyplot as plt

# 读取 JSON 文件
with open('latency.json', 'r') as f:
    data = json.load(f)

# 检查是否成功加载数据
if not data:
    print("JSON 数据为空，请检查生成 JSON 的过程。")
    exit(1)

# 创建图形和子图
plt.figure(figsize=(10, 6))

# 遍历 JSON 中的每个主题，并绘制延迟曲线
for topic, latencies in data.items():
    # 构造 x 轴数据：消息的索引
    x = list(range(1, len(latencies) + 1))
    plt.plot(x, latencies, marker='o', label=topic)

# 设置图表标题和标签
plt.title('各主题消息延迟（毫秒）')
plt.xlabel('消息序号')
plt.ylabel('延迟 (ms)')
plt.legend()
plt.grid(True)

# 显示图表
plt.show()
```
